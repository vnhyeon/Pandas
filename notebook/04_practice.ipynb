# 앤스콤 데이터 집합 불러오기

```python
# seaborn은 나름 고급 그래프 라이브러리
import seaborn as sns

# 4개의 그룹으로 구성되어 있는 anscombe data 사용
anscombe = sns.load_dataset("anscombe") 

print(anscombe)
print(type(anscombe))
```

# matplotlib 라이브러리로 간단한 그래프 그리기

```python
# 데이터 집합을 불러왔으니 그래프로 시각화
import matplotlib.pyplot as plt

# 데이터 값이 I인 것만 추출
dataset_1 = anscombe[anscombe['dataset'] == 'I'] 

# x, y축 데이터를 plot 메서드에 전달
plt.plot(dataset_1['x'], dataset_1['y'])

# 점으로 그래프를 그리고 싶다면 'o'
plt.plot(dataset_1['x'], dataset_1['y'], 'o') 

# 위의 그래프를 출력하는 명령어. 유사print의 역할
plt.show() 
```

# 한 번에 4개의 그래프 그리기

```python
dataset_1 = anscombe[anscombe['dataset'] == 'I']
dataset_2 = anscombe[anscombe['dataset'] == 'II']
dataset_3 = anscombe[anscombe['dataset'] == 'III']
dataset_4 = anscombe[anscombe['dataset'] == 'IV']

# 그래프 격자가 위치할 기본 틀 제작
fig = plt.figure() 

# add_subplot 메서드로 그래프 틀인 fig에 그래프 격자 부여
axes1 = fig.add_subplot(2, 2, 1) 
axes2 = fig.add_subplot(2, 2, 2)
axes3 = fig.add_subplot(2, 2, 3)
axes4 = fig.add_subplot(2, 2, 4)

# plot 메서드에 데이터 전달
axes1.plot(dataset_1['x'], dataset_1['y'], 'o')
axes2.plot(dataset_2['x'], dataset_2['y'], 'o')
axes3.plot(dataset_3['x'], dataset_3['y'], 'o')
axes4.plot(dataset_4['x'], dataset_4['y'], 'o')

# suptitle 메서드로 fig 제목 부여
fig.suptitle("anscombe Data")

# set_title 메서드로 그래프 제목 부여
axes1.set_title("dataset_1") 
axes2.set_title("dataset_2")
axes3.set_title("dataset_3")
axes4.set_title("dataset_4")

# .tight_layout 메서드를 통해 각 그래프의 레이아웃 자동 조절
fig.tight_layout() 

```

# 기초 그래프 그리기

```python
### 히스토그램
# seaborn 라이브러리에는 tips라는 데이터 집합이 존재. 실제 식당에서의 tips임

# total_bill만을 변수로 하는 일변량 그래프 그릴거야
tips = sns.load_dataset("tips") 
print(tips.head())
print(type(tips))

fig = plt.figure()
axes1 = fig.add_subplot(1, 1, 1)

axes1.set_title('Histogram of Total Bill')
axes1.set_xlabel('Frequency')
axes1.set_ylabel('Total Bill')

# hist 메서드에 변수열을 전달, bins 인잣값은 x축의 간격을 의미
axes1.hist(tips['total_bill'], bins=10) 

```

```python
### 산점도
# 2개의 변수를 사용한 이변량 그래프 (x, y) = (Total Bill, Tip)

scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1, 1, 1)

axes1.set_title('Scatterplot of Total Bill vs Tip')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tips')

# scaater 메서드에 x, y 순서대로 tips['전달할 열']를 전달
axes1.scatter(tips['total_bill'], tips['tip']) 
```

```python
### 박스 그래프
# 이산형 변수(주로x)와 연속형 변수(주로y)를 함께 사용하는 그래프

boxplot = plt.figure()
axes1 = boxplot.add_subplot(1, 1, 1)

axes1.set_title('Boxplot of Tips by Sex')
axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')

# tips df에서 tips 시리즈의 성별이 Female인 행만 필터링하고 맨 뒤의 ['tip']으로 tip 열만 선택한 것.
axes1.boxplot([tips[tips['sex'] == 'Female']['tip'], 
               tips[tips['sex'] == 'Male']['tip']],
               labels=['Female', 'Male'])
```

# 다변량 그래프 그리기

```python
# 3개 이상의 변수를 사용한 그래프
# 위에서 그린 산점도에서 성별을 추가하려면 점의 색상, 점의 크기 등으로 변수를 추가 할 수 있음
# 성별은 색상으로 구분 가능하지만 Female, Male 같은 문자열은 색상으로 사용할 수 없음
# 따라서 0, 1과 같은 정수를 사용해야하고 문자열을 정수로 치환하는 함수를 만들 것

def recode_sex(sex):
    if sex == 'Female':
        return 0
    else:
        return 1

# 위의 정수를 시리즈로 추가하며 def 함수를 브로드캐스팅 하기 위해 apply 메서드 사용    
tips['sex_color'] = tips['sex'].apply(recode_sex) 

scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1, 1, 1)

axes1.set_title('Total Bill vs Tip Colored by sex and Sized by size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')

axes1.scatter(
    x=tips['total_bill'],
    y=tips['tip'],
    s=tips['size']*10, # 테이블 당 인원 수는 점의 크기를 늘린것 
    c=tips['sex_color'], # 성별의 치환 값 전달
    alpha=0.5) # 점의 투명도를 보기좋게 조절한 것
```

# seaborn 라이브러리 자유자재로 사용하기

```python
### 히스토그램
# seaborn 라이브러리는 matplotlib 라이브러리를 기반으로 만든 것

import seaborn as sns
tips = sns.load_dataset("tips")

# subplot 메소드로 기본 틀을 만든 것
ax = plt.subplot() 

# distplot(사용하고 싶은 열 변수를 전달하는 메서드)이 권장하지 않는 구식의 방법이라 선이 있는것
ax = sns.distplot(tips['total_bill']) 
# 밀집도 그래프를 제외하고 싶으면 ax = sns.distplot(tips['total_bill'], kde=False)
# 밀집도 그래프만 나타내려면 ax = sns.distplot(tips['total_bill'], hist=False)

ax.set_title('Total Bill Histogram with Density Plot')

# rug=True 입력 시 양탄자 그래프까지 추가 돼. 좀 더 실무 특화
ax = sns.distplot(tips['total_bill'], rug=True) 
```

```python
### count 그래프
ax = plt.subplots()

# countplot 메서드에 tips 시리즈를 할당
ax = sns.countplot(x = 'day', data = tips) 
```

```python
### 산점도
# seaborn 라이브러리의 regplot 메서드를 사용하면 돼

ax = plt.subplot()
ax = sns.regplot(x='total_bill', y='tip', data=tips)
# 회귀선을 없애려면 ax = sns.regplot(x='total_bill', y='tip', data=tips, fig_reg=False)
```

```python
### 산점도와 히스토그램
# jointplot 메서드 사용
joint = sns.jointplot(x='total_bill', y='tip', data=tips) # 축에 변수 할당

joint.set_axis_labels(xlabel='Total Bill', ylabel='Tip') # 축 제목 할당

joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize=10, y=1.03) # 전체 제목 할당
```

```python
### 육각 그래프
# 산점도 그래프의 데이터를 구분하기 쉽게 그리는 도구

joint = sns.jointplot(x='total_bill', y='tip', data=tips, kind="hex")
#위의 문법과 같고 kind="hex" 추가만 해준 경우
```

```python
### 이차원 밀집도
# kdeplot 메서드로 그릴 수 있음
ax = plt.subplot()
ax = sns.kdeplot(x=tips['total_bill'],
                 y=tips['tip'],
                 shade=True) # shade=True는 음영효과
```

```python
### 바, 박스 그래프
# 바 그래프는 barplot
ax = plt.subplot()
ax = sns.barplot(x='time', y='total_bill', data=tips)

# 박스 그래프는 boxplot
ax = plt.subplot()
ax = sns.boxplot(x='time', y='total_bill', data=tips)
```

```python
### 바이올린, 관계 그래프
# 박스 그래프 + 커널 밀도 추정한 그래프 violinplot 메서드 사용
ax = plt.subplot()
ax = sns.violinplot(x='time', y='total_bill', data=tips)

# 위의 모든 그래프를 종합한 그래프 pairplot 메서드에 df를 넣는 방법으로 그림
fig = sns.pairplot(tips)

# pairplot 그래프는 주대각선 기준으로 대칭인 단점이 있음. 이걸 메소드로 지정해줘
pair_grid = sns.PairGrid(tips)
pair_gird = pair_grid.map_upper(sns.regplot) # 대각선 기준 위쪽
pair_gird = pair_grid.map_lower(sns.kdeplot) # 대각선 기준 아래쪽
pair_gird = pair_grid.map_diag(sns.distplot, rug=True) # 주대각선 기준
plt.show()
```

# seaborn 라이브러리로 그래프 스타일 설정하기

```python
# seaborn 라이브러리로 만든 위 그래프의 스타일은 set_style 메서드로 바꿀 수 있어
# 실행한 이후 그래프가 모두 지정한 스타일로 바뀌어
# darkgrid, whitegrid, dark, white, ticks 5종류가 존재

sns.set_style('whitegrid') # 배경에 가로줄 삽입
ax = plt.subplot()
ax = sns.violinplot(x='time', y='total_bill', hue='sex', data=tips, split=True)

fig = plt.figure()
seaborn_styles = ['darkgrid', 'whitegrid', 'dark', 'white', 'ticks']

# enumerate가 idx와 style을 함께 반환
for idx, style in enumerate(seaborn_styles): 
    plot_position = idx + 1
    with sns.axes_style(style): # with 블록 안에는 모두 여기서 선언한 style로 변경
        ax = fig.add_subplot(2, 3, plot_position) # 2행 3열의 그리드에서 position에 맞는 서브플롯 추가
        violin = sns.violinplot(x='time', y='total_bill', data=tips, ax=ax)
        violin.set_title(style)
        
fig.tight_layout()
```
